```python
!pip install requests
!pip install beautifulsoup4
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_programming_languages'
response = requests.get(url)

if response.status_code == 200:
    print('Successfully fetched the web page')
else:
    print('Failed to fetch the web page', response.status_code)

soup = BeautifulSoup(response.content, 'html.parser')

content = soup.find_all('div', class_='div-col')

languages = []
for div in content:
    for li in div.find_all('li'):
        languages.append(li.text)

for language in languages:
    print(language)

df = pd.DataFrame(languages, columns=['Programming Language'])
df.to_csv('programming_languages.csv', index=False)
```

```python
!pip install requests
!pip install beautifulsoup4
!pip install pandas
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def scrape_yelp_business_data(page_url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    response = requests.get(page_url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    businesses = []

    for business in soup.find_all('div', class_='container__09f24__21w3G'):
        name = business.find('a', class_='link__09f24__1MGLa').text.strip() if business.find('a', class_='link__09f24__1MGLa') else None
        address = business.find('address', class_='raw__09f24__3Azrj').text.strip() if business.find('address', class_='raw__09f24__3Azrj') else None
        phone = business.find('p', class_='text__09f24__2NHRu').text.strip() if business.find('p', class_='text__09f24__2NHRu') else None

        if name and address and phone:
            businesses.append({
                'Name': name,
                'Address': address,
                'Phone': phone
            })

    return businesses

def scrape_multiple_pages(base_url, start_page, end_page):
    all_businesses = []

    for page in range(start_page, end_page + 1):
        page_url = f"{base_url}&start={page*10}"
        businesses = scrape_yelp_business_data(page_url)
        all_businesses.extend(businesses)
        time.sleep(2)  # Pause to avoid being blocked

    return all_businesses

def save_to_csv(businesses, filename):
    df = pd.DataFrame(businesses)
    df.to_csv(filename, index=False)

# Base URL of the Yelp directory (for example, searching for restaurants in San Francisco)
base_url = 'https://www.yelp.com/search?find_desc=restaurants&find_loc=San+Francisco%2C+CA'

# Scrape data from the first 5 pages
business_data = scrape_multiple_pages(base_url, start_page=0, end_page=5)

# Save the data to a CSV file
save_to_csv(business_data, 'b2b_leads_yelp.csv')

# Display the data
df = pd.read_csv('b2b_leads_yelp.csv')
df.head()
```
